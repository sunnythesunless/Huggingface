```python
from transformers import pipeline
import gradio as gr

sentiment = pipeline("sentiment-analysis", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
generator = pipeline("text-generation", model="gpt2")
qa = pipeline("question-answering", model="distilbert/distilbert-base-cased-distilled-squad")
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
ner = pipeline("ner", model="dslim/bert-base-NER")

print("✅ All models loaded!")
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Hugging face is awsome!"
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

print(f"original: {text}")
print(f"token: {tokens}")
print(f"token_ids: {token_ids}")
```

```python
text2 = "Hello my name is gonebone"
tokens2 = tokenizer.tokenize(text2)
print(f"\nTokens for '{text2}': {tokens2}")
```

```python
# --- 1. Sentiment ---
result = sentiment("I love building AI projects, this is exciting!")
print("Sentiment:", result)


result2 = sentiment("this fucking comptition all over world hard to fucking get a job")
print("Sentiment:", result2)
```

```python
# --- 2. Text Generation ---
result = generator("The future of AI is", max_new_tokens=30, num_return_sequences=1)
print(result[0]["generated_text"])
```

```python
# 3. Q&A
print("\n--- QUESTION ANSWERING ---")
result = qa(question="When was HF founded?", context="Hugging Face was founded in 2016 in New York.")
print(f"A: {result['answer']} (confidence: {result['score']:.2f})")
```

```python
##4. zero shot
print("\n--- ZERO-SHOT ---")
result = classifier("The stock market crashed today", candidate_labels=["finance", "sports", "tech"])
print(list(zip(result["labels"], [f'{s:.2f}' for s in result["scores"]])))
```

```python
# Re-load the NER pipeline with the aggregation strategy
ner = pipeline("ner", aggregation_strategy="simple")
```

```python
result = ner("Elon Musk founded SpaceX in California.")

for entity in result:
    print(f"{entity['word']}: {entity['entity_group']} (confidence: {entity['score']:.2f})")
```

```python
def analyze_sentiment(text):
    r = sentiment(text)[0]
    emoji = "😊" if r["label"] == "POSITIVE" else "😞"
    return f"{emoji} {r['label']} — {r['score']:.1%}"

def generate_text(prompt):
    r = generator(prompt, max_length=80, num_return_sequences=1)
    return r[0]["generated_text"]

def answer_question(context, question):
    r = qa(question=question, context=context)
    return f"Answer: {r['answer']}\nConfidence: {r['score']:.1%}"

def classify_text(text, labels):
    label_list = [l.strip() for l in labels.split(",")]
    r = classifier(text, candidate_labels=label_list)
    output = ""
    for label, score in zip(r["labels"], r["scores"]):
        bar = "█" * int(score * 30)
        output += f"{label}: {score:.1%} {bar}\n"
    return output

def detect_entities(text):
    results = ner(text)
    output = ""
    for e in results:
        output += f"{e['word']} → {e['entity_group']} (confidence: {e['score']:.1%})\n"
    return output if output else "No entities found."

with gr.Blocks(title="🤗 AI Toolkit", theme=gr.themes.Soft()) as app:
    gr.Markdown("# 🤗 Multi-Task AI Toolkit")
    gr.Markdown("*5 AI tasks powered by Hugging Face Transformers*")

    with gr.Tab("😊 Sentiment"):
        s_in = gr.Textbox(label="Text", placeholder="Type anything...")
        s_out = gr.Textbox(label="Result")
        gr.Button("Analyze", variant="primary").click(analyze_sentiment, s_in, s_out)

    with gr.Tab("✍️ Generate"):
        g_in = gr.Textbox(label="Prompt", placeholder="Start a sentence...")
        g_out = gr.Textbox(label="Generated Text", lines=4)
        gr.Button("Generate", variant="primary").click(generate_text, g_in, g_out)

    with gr.Tab("❓ Q&A"):
        qa_ctx = gr.Textbox(label="Context", lines=4, placeholder="Paste a paragraph...")
        qa_q = gr.Textbox(label="Question", placeholder="Ask something about the context...")
        qa_out = gr.Textbox(label="Answer")
        gr.Button("Answer", variant="primary").click(answer_question, [qa_ctx, qa_q], qa_out)

    with gr.Tab("🏷️ Classify"):
        c_in = gr.Textbox(label="Text", placeholder="Any sentence...")
        c_labels = gr.Textbox(label="Categories (comma-separated)", placeholder="sports, finance, tech")
        c_out = gr.Textbox(label="Results", lines=4)
        gr.Button("Classify", variant="primary").click(classify_text, [c_in, c_labels], c_out)

    with gr.Tab("🔍 NER"):
        n_in = gr.Textbox(label="Text", placeholder="e.g. Elon Musk founded SpaceX in California")
        n_out = gr.Textbox(label="Entities Found", lines=4)
        gr.Button("Detect", variant="primary").click(detect_entities, n_in, n_out)

app.launch(share=True)
```
